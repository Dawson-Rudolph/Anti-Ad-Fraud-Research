```python
import pandas as pd
import numpy as np
import data_reprocessing as rp
import training_preprocess as tp




#for testing
#data = tp.pre_process_train('train_sample.csv')

#for undersampling
data = rp.train_to_df('new_train.csv')
```

```python
#refactor dataframe into two columns of fraud and non fraud
#remove a LARGE portion of fraud as you go

data_new = {0 : [], 1 : []}

for index, row in data.iterrows():

    if row['is_attributed'] == 0:
        if np.random.randint(1, 1000) > 990:
            data_new[row['is_attributed']].append([row['ip'], row['app'], row['device'], row['os'], row['channel'], row['day'], row['hour'], row['minute'], row['second']])
    else:
        data_new[row['is_attributed']].append([row['ip'], row['app'], row['device'], row['os'], row['channel'], row['day'], row['hour'], row['minute'], row['second']])

#test split for good numbers
print(f'fraud : benign -- {len(data_new[0])/(len(data_new[0])+len(data_new[1]))} : {len(data_new[1])/(len(data_new[0])+len(data_new[1]))}')
```

```python
#dictionary to refit undersampled data to original structure
data_refit = {col : [] for col in data.columns}

for row in data_new[0]:
    data_refit['ip'].append(row[0])
    data_refit['app'].append(row[1])
    data_refit['device'].append(row[2])
    data_refit['os'].append(row[3])
    data_refit['channel'].append(row[4])
    data_refit['is_attributed'].append(0)
    data_refit['day'].append(row[5])
    data_refit['hour'].append(row[6])
    data_refit['minute'].append(row[7])
    data_refit['second'].append(row[8])

for row in data_new[1]:
    data_refit['ip'].append(row[0])
    data_refit['app'].append(row[1])
    data_refit['device'].append(row[2])
    data_refit['os'].append(row[3])
    data_refit['channel'].append(row[4])
    data_refit['is_attributed'].append(1)
    data_refit['day'].append(row[5])
    data_refit['hour'].append(row[6])
    data_refit['minute'].append(row[7])
    data_refit['second'].append(row[8])

balanced = pd.DataFrame(data_refit)
balanced.head()
```

```python
balanced.to_csv('balanced_train.csv', index=False)
```
